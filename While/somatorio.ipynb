```python
n = 10  

contador = 0
soma = 0
numero_impar_atual = 1


while contador < n:
        soma += numero_impar_atual
        numero_impar_atual += 2
        contador += 1
    
print("O total de números impares foi {}.".format(contador))
print("A soma dos números impares foi {}.".format(soma))
```
