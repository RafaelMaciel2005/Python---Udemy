```python
"""" faça um programa que peça a senha de 4 digitos para o usuario e confirme que está correta"""
senha = int(input("Digite uma senha de 4 digitos:"))
senha_real = 1234

while senha != senha_real:
    print("Senha incorreta, tente novamente.")
    senha = int(input("Digite uma senha de 4 digitos:"))
if senha == senha_real:
    print("Senha correta.")        
```
